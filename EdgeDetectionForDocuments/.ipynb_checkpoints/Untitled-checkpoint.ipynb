```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread("blackBackground.JPG")
plt.imshow(image)
plt.show()
```

```python
grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(grey)
plt.show()
```

```python
blur=cv2.GaussianBlur(grey,(5,5),1)
plt.imshow(blur)
plt.show()
```

```python
def auto_canny(image, sigma=0.33):
   
    v = np.median(image)
   
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    edged = cv2.Canny(image, lower, upper)
    
    return edged
```

```python
edges=auto_canny(blur)
plt.imshow(edges)
plt.show()
```

```python
imgContours=image.copy()

contours,hierarchy1=cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[-2:]

image_contours=cv2.drawContours(imgContours,contours,-1,(0,255,0),10)

plt.imshow(image_contours)
plt.show()
```

```python
def biggestContour(contours):
    biggest=np.array([])
    max_area=0
    for i in contours:
        area=cv2.contourArea(i)
        if area>5000:
            peri=cv2.arcLength(i,True)
            approx=cv2.approxPolyDP(i,0.02*peri,True)
            if area > max_area :
                biggest=approx
                max_area=area
    return biggest,max_area
    
```

```python
biggest,max_area=biggestContour(contours)

print(biggest)
```

```python
imgContours=image.copy()

biggest_contour=cv2.drawContours(imgContours,biggest,-1,(0,255,0),10)

plt.imshow(biggest_contour)
plt.show()
```
